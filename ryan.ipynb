```python
import pandas as pd
```

```python
fooddata = "Resources/DataDownload2015.xlsx"
```

```python
food = pd.read_excel(fooddata, sheet_name="Food Access Research Atlas")
```

```python
food.head()
```

```python
foodf = food
foodf['poverish_people_per'] = 100* ((foodf.POP2010 * (foodf.PovertyRate/100))/foodf['Total Population'])
foodf.groupby('County').agg({'poverish_people_per': 'sum'})
```

```python
foodf.head()
```

```python
caltemp = foodf.loc[foodf["State"] == "California",:]
cal = caltemp.groupby('County').agg({'poverish_people_per': 'sum'})
```

```python
caltemp.set_index("County")
caltemp.head()
```

```python
ccounty = caltemp.drop_duplicates("County")
counties = ccounty[["County", "Total Population"]]
counties.head()
counties.set_index("County")
```

```python
counties.head()
```

```python
test = cal[["poverish_people_per"]]
```

```python
test.head()
```

```python
final = test.merge(counties, how='left', on='County')
final.head()
```

```python
fin = final.rename(columns = {"poverish_people_per" : "Poverty Rate"})
```

```python
fin = fin.round(2)
```
