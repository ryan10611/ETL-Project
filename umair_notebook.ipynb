```python

import pandas as pd
```

```python
#Falling Fruit Data from:
# https://www.fallingfruit.org/data?c=forager%2Cfreegan&locale=en
#used BZ2 to ZIP converter https://cloudconvert.com/bz2-to-zip
```

```python
#import types data
types_df = pd.read_csv('Resources/types.csv')
```

```python
#there are dumpsters listed on there and we don't want those
types_df = types_df[types_df['en_name'] != 'Dumpster (edible)']
```

```python
##select fewer columns
types_df = types_df[['id', 'parent_id', 'scientific_name', 'en_name', 'en_wikipedia_url']]
types_df = types_df.rename(columns={'id': 'type_ids'})
```

```python
#change type for merge
types_df['type_ids'] = types_df['type_ids'].astype('str')
```

```python
# types_df.to_csv('Resources/types.csv')
#the original file is too big for GitHub
```

```python
#this file is huge, with data for the entire world
locations_df = pd.read_csv('Resources/locations.csv', low_memory=False)
```

```python
#select necessary columns
locations_df = locations_df[['id', 'type_ids', 'lat', 'lng', 'access']]
```

```python
#get latitude and longitude
locations_df = locations_df.rename(columns={'lng': 'lon'})
```

```python
locations_df = locations_df[(locations_df['lat'].round(decimals=0).astype('int') > 32)]
locations_df = locations_df[(locations_df['lat'].round(decimals=0).astype('int') <= 42)]
locations_df = locations_df[(locations_df['lon'].round(decimals=0).astype('int') < -114)]
locations_df = locations_df[(locations_df['lon'].round(decimals=0).astype('int') >= -125)]
```

```python
#export my cleaned data
locations_df.to_csv('Resources/locations.csv')
# locations_df.head()
```

```python
#https://geopy.readthedocs.io/en/stable/
#need to pip install geopy
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="foraging_addresses")
```

```python
#This is an example to show what the function above returns
location = geolocator.reverse("30.294962, -97.744891")
location.raw
```

```python
coords = locations_df['lat'].astype('str').str.cat(locations_df['lon'].astype('str'), sep=", ")
coords
```

```python
# Due to limit of 1291 on requests, we do not want to run this too many times
locations = []
for row in coords:
    locations.append(geolocator.reverse(row))
```

```python
#turn result into list
addresses = []
for i in locations:
    addresses.append(i.raw)
```

```python
address_df = pd.DataFrame(addresses)
```

```python
#you can see there are lists and dictionaries in this database
address_df.head()
```

```python
#get latitude and longitude
address_df[['lat', 'lon']].head()
```

```python
address_df['address'].apply(pd.Series)
```

```python
address_df_trim = pd.concat([address_df[['lat', 'lon']], address_df['address'].apply(pd.Series)], axis=1)
```

```python
address_df_trim = address_df_trim[['lat', 'lon', 'house_number',	'road',	'neighbourhood', 'city',	'county',	'state', 	'postcode',	'country']]
```

```python
fallingfruit_df = locations_df.merge(types_df, how='inner', on='type_ids')
```

```python
#the resultant data retained the order and therefore index of what we sent in 
fallingfruit_df.merge(address_df_trim, left_index=True, right_index=True)
```

```python
#want to limit data to CA
final_df = address_df_trim[address_df_trim['state']=='California']
```
